### Parse NCBI Taxonoy

```python
import pandas as pd
import networkx as nx
```

```python
tree_path = './taxid_to_parent.names.tsv'

matrix = '../uniprotkb_reference_proteomes/big_bacteria_countmatrix.csv'
```

```python
matrix_df = pd.read_csv(matrix, index_col=0)
matrix_df
```

```python
tax_tree = nx.DiGraph()
with open(tree_path) as f:
	for line in f:
		taxid, parent, name = line.strip().split('\t')
		tax_tree.add_node(taxid, name=name)
		if taxid != parent:
			tax_tree.add_edge(parent, taxid)
```
